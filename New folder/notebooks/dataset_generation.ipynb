```python
import pandas as pd
import numpy as np

```

```python
n = 5000
```

```python
np.random.seed(42)

data = {
    "area_type": np.random.choice(["Residential", "Commercial", "Public"], n),
    "time_of_day": np.random.choice(["Morning", "Evening", "Night"], n),
    "day_of_week": np.random.choice(["Weekday", "Weekend"], n),
    "past_incidents": np.random.randint(0, 20, n),
    "police_presence": np.random.choice(["Low", "Medium", "High"], n),
    "lighting": np.random.choice(["Poor", "Average", "Good"], n),
    "crowd_density": np.random.choice(["Low", "Medium", "High"], n)
}
```

```python
risk = []

for i in range(n):
    score = 0
    
    if data["past_incidents"][i] > 10:
        score += 2
    if data["time_of_day"][i] == "Night":
        score += 1
    if data["lighting"][i] == "Poor":
        score += 1
    if data["police_presence"][i] == "Low":
        score += 1
    
    if score >= 4:
        risk.append(2)   # High
    elif score >= 2:
        risk.append(1)   # Medium
    else:
        risk.append(0)   # Low
```
