```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Jupyter Notebook magic
%matplotlib inline

# Load dataset
df = pd.read_csv("women_safety_data.csv")
df.head()
```

```python
sns.countplot(x='risk_level', data=df)
plt.title("Risk Level Distribution")
plt.show()
```

```python
categorical_cols = ["area_type", "time_of_day", "day_of_week", 
                    "police_presence", "lighting", "crowd_density"]

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, hue='risk_level', data=df)
    plt.title(f"{col} vs Risk Level")
    plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['past_incidents'], bins=20, kde=True)
plt.title("Past Incidents Distribution")
plt.show()

sns.boxplot(x='risk_level', y='past_incidents', data=df)
plt.title("Past Incidents vs Risk Level")
plt.show()
```
